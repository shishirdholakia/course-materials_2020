## Intro to Stellar Rotation
### AY 128/256 (UC Berkeley, 2020)

* Stars are not static objects
* Lots of motion and activity
 * rotation, star spots, flares, convective cell transport, ...
* There processes are observable through sensitive time series data (spectra and/or photometry)
 * The typical effect is pretty small, often $<<$1\%
 * challenging to get this level of precision from ground-based telescopes (e.g,. our atmosphere varies too much)
  * Time series + sensitivity: telescopes like Kepler and TESS have been excellent sources for understanding the detailed physics of stars
  

### Stellar Rotation
* Stars rotate
 - causes oblation -- stars are more elongated around the equator than around the poles
 - rotation can cause small doppler shifts in the spectrum of a star, unless it is observed pole-on
 
* Stars rotate differentially
 - the interaction of convective heat transport and rotation produces differential rotation and differential magnetic fields
 - these pheneoma can induce differential rotation in a star such that rotation at a star's equator can be up to twice as fast as at higher latitudes
 - these effects can also induce star spots, which appear as hot and cold patches on the surface of a star
 - tracking location of star spots as a function of time is a way to measure rotation and differential rotation
 
* Stellar Activity (such as star spots)
 - complicated -- hard to model from first principles
 - many mechanisms produce similar observable effects (e.g., variations in brightness can be due to rotation, star spots, stellar flares, transiting planets)
 - can be challenging to decouple these effects

* Data driven model (e.g,. Gaussian processes) provides a way to extract information about the rotation of a star even in absence of a good physical model
