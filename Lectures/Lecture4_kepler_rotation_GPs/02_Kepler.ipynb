## Kepler Space Telescope

### AY 128/256 (UC Berkeley, 2020)

<img src="https://i.insider.com/5a4d2ea4cf698afe0e8b48f4?width=2500&format=jpeg&auto=webp"></img>


* Main science goal was to find Earth-like exoplanets in the habitable zones of nearby Sun-like stars by looking for dips in the brightness of stars (more on this below)

* Total mission duration May 7, 2009 - Oct 30, 2018
 * One of the 4 reaction wheels needed to stablize the telescope failed in 2012
 * Another reactin wheel failed in 2013
 * Had to re-configure the main purpose of the mission given the failure of the reaction wheels
 * Initially planned as 3.5 year mission; "K2" extended the mission for another ~6 years
  * K2 shifted focus to planets around cooler, dimmer stars 


* Telescope characteristics
 * 1.4m primary mirror
  * wide-field: 105 deg$^2$ field of view
  * designed to detect brightness fluctuations at the level of $2\times10^{-5}$ for a 12th magnigutde of a Sun-like in ~6 hours


* Kepler field of view

<img src="https://supernovacondensate.files.wordpress.com/2012/07/kepler-set-7.jpg"></img>


* Kepler soley did imaging, looking for dips in the brightness of nearby stars to indicate the presence of a transiting planet
<img src="https://www.nasa.gov/sites/default/files/styles/full_width_feature/public/aas2010-1wblightcurves2-full.jpg"></img>


* Many sources of noise in the light curve of than planet transits
 * stellar activity -- e.g., rotation, flares, star spots, etc.
 * Instrumental noise -- e.g., differential velocity aberration ($\sim$ 1\%; from a given star moving to a different location on the CCD over the course of time), changes in the thermal properties ($\sim$0.1\% effect), shot and correlated noise
